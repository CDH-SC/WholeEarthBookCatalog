# Whole Earth Book Catalog : The Data
The data files are organized into separate node files, edge/relationship files, and combined files
## Nodes
There are four different types of nodes: **edition, person, place, and publisher** <br>
- combined_batch.tsv
     - **File size: 1.9GB**
     - Concatenation of below files with columns: ID, Name, Type, ISBN, and Year
     - The files contain repeated IDs
         - for example: Node with ID, 3 and Name, Paris can be found repeated in the table
     - Column titles not included in file
